# Solve for Galactocentric radial velocity given line of sight velocity, distance, Galactic coordinates, and an assumed circular velocity in the disk

The observed quantities:

- Line of sight velocity
- Distance
- Galactic coordinates

Assumed quantities:

- Circular velocity within the disk
- Sun's position and motion relative to the Galactic center and local standard of rest

First, it is helpful to convert everything to Galactocentric cylindrical coordinates. The observed quantities are in heliocentric spherical coordinates. The conversion:

$$
\begin{align}
\langle D, l, b &\rangle \rightarrow \langle D \cos l \cos b, C \sin l \cos b, D \sin b \rangle
\\
&\rightarrow \langle D \cos l \cos b - R_\odot, D \sin l \cos b, D \sin b \rangle
\\
&\rightarrow \langle \sqrt{(D \cos l \cos b - R_\odot)^2 + D^2 \sin^2 l \cos^2 b}, \phi, D \sin b \rangle,
\end{align}
$$

where $\phi$ is the Galactocentric angle in cyllindrical coordinates, given by

$$
\phi = \tan^{-1}\left(\frac{D^2 \sin^2 l \cos^2 b}{\sqrt{(D \cos l \cos b - R_\odot)^2 + D^2 \sin^2 l \cos^2 b}}\right),
$$

but it turns out to be more useful to solve for $\sin \phi$ and $\cos \phi$ directly:

$$
\begin{align}
\sin \phi &= \frac{y}{R_c} = \frac{D \sin l \cos b}{
    \sqrt{(D \cos l \cos b - R_\odot)^2 + D^2 \sin^2 l \cos^2 b}
}
\\
\cos \phi &= \frac{x}{R_c} = \frac{D \cos l \cos b - R_\odot}{
    \sqrt{(D \cos l \cos b - R_\odot)^2 + D^2 \sin^2 l \cos^2 b}
},
\end{align}
$$

where $R_c$ is the Galactocentric radius of the HI cloud.

Now, in Galactocentric cylindrical coordinates, we can relate the distance and relative velocities of the Sun and the observed HI cloud to the observed quantities. The vector displacement between the two:

$$
\begin{align}
\vec r_\odot &= \langle -R_\odot, 0, z_\odot \rangle
\\
\vec r_c &= \langle R_c \cos \phi, R_c \sin \phi, z_c \rangle
\\
\implies \vec r_c - \vec r_\odot &= \langle R_c \cos \phi + R_\odot, R_c \sin \phi, z_c - z_\odot \rangle.
\end{align}
$$

The relative velocity vector is simply the time derivative of the displacement vector:

$$
\begin{align}
\vec v_\odot &= \langle v_{\odot,x}, v_{\odot,y}, v_{\odot,z} \rangle = \vec v_\text{circ} + \vec v_\text{pec}
= \langle U, V - v_\text{circ}, W \rangle
\\
\vec v_c &= \langle \dot R_c \cos \phi - R_c \dot \phi \sin \phi, \dot R_c \sin \phi + R_c \dot \phi \cos \phi, \dot z_c \rangle
\\
&= \langle v_r \cos \phi - v_\phi \sin \phi, v_r \sin \phi + v_\phi \cos \phi, v_z \rangle
\\
\implies \vec v_c - \vec v_\odot &= \langle
v_r \cos \phi - v_\phi \sin \phi - v_{\odot,x},
v_r \sin \phi + v_\phi \cos \phi - v_{\odot,y},
v_z - v_{\odot,z},
\rangle
\end{align}
$$

where we have swapped out $\dot R_c = v_r$ for the radial velocity (our primary quantity of interest), along with the azimuthal velocity in the plane of the disk $R \dot \phi = v_\phi$ and the velocity perpendicular to the disk $\dot z_c = v_z$.

The measured line of sight velocity is the magnitude of the relative velocity projected onto the displacement vector:

$$
\begin{align}
v_\text{los} &= \left| \text{proj}_{\vec r_c - \vec r_\odot} (\vec v_c - \vec v_\odot) \right|
\\
&= \frac{1}{D} \left(\vec v_c - \vec v_\odot\right) \cdot \left(\vec r_c - \vec r_\odot\right)
\\
\implies D v_\text{los} &=
\left(v_r \cos\phi - v_\phi \sin\phi - v_{\odot,x}\right)\left(R_c \cos\phi + R_\odot\right) +
\left(v_r \sin\phi + v_\phi \cos\phi - v_{\odot,y}\right) R_c \sin\phi +
\left(v_z - v_{\odot,z}\right)\left(z_c - z_\odot\right)
\\
&= v_r \cos\phi \left(R_c \cos\phi + R_\odot\right) + v_r R_c \sin^2 \phi -
\left(v_\phi \sin\phi + v_{\odot,x}\right)\left(R_c \cos\phi + R_\odot\right) +
\left(v_\phi \cos\phi - v_{\odot,y}\right) R_c \sin\phi +
\left(v_z - v_{\odot,z}\right)\left(z_c - z_\odot\right)
\\
&= v_r \left(R_c + R_\odot \cos\phi\right) -
\left(v_\phi \sin\phi + v_{\odot,x}\right)\left(R_c \cos\phi + R_\odot\right) +
\left(v_\phi \cos\phi - v_{\odot,y}\right) R_c \sin\phi +
\left(v_z - v_{\odot,z}\right)\left(z_c - z_\odot\right)
\\
\implies v_r &= \frac{
    D v_\text{los}
    + \left(v_\phi \sin\phi + v_{\odot,x}\right)\left(R_c \cos\phi + R_\odot\right)
    - \left(v_\phi \cos\phi - v_{\odot,y}\right) R_c \sin\phi
    - \left(v_z - v_{\odot,z}\right)\left(z_c - z_\odot\right)
}{
    R_c + R_\odot \cos\phi
}
\end{align}
$$

The vertical components of the line of sight velocity and displacement vector can both be written in terms of the distance, velocity, and Galactic latitude as $v_z - v_{\odot,z} = v_\text{los} \sin b$ and $z_c - z_\odot = D \sin b$.
Therefore we can make one final substitution:

$$
\begin{align}
v_r &= \frac{
    D v_\text{los}
    + \left(v_\phi \sin\phi + v_{\odot,x}\right)\left(R_c \cos\phi + R_\odot\right)
    - \left(v_\phi \cos\phi - v_{\odot,y}\right) R_c \sin\phi
    - D v_\text{los} \sin^2 b
}{
    R_c + R_\odot \cos\phi
}
\\
&= \frac{
    D v_\text{los} \cos^2 b
    + \left(v_\phi \sin\phi + v_{\odot,x}\right)\left(R_c \cos\phi + R_\odot\right)
    - \left(v_\phi \cos\phi - v_{\odot,y}\right) R_c \sin\phi
}{
    R_c + R_\odot \cos\phi
},
\end{align}
$$

and we now arrive at an expression for the radial velocity that is directly solvable from our measured quantities under the assumption of some azimuthal velocity vector $v_\phi$.

```python
import matplotlib.pyplot as plt
from plots.mpltoolkit import mpl_loc, named_colors, fancy_legend, markers, load_style_sheet
import numpy as np
import math as m
import vice
import sys
load_style_sheet("papers")
print(sys.version_info)
```

```python
class hi_cloud:

    KPC_TO_KM = 3.086e16
    # RSUN = 8.5 * KPC_TO_KM
    RSUN = 8 * KPC_TO_KM
    VCIRC = 220 # km/s
    U = 0 # km/s
    V = 0 # km/s
    W = 0 # km/s

    def __init__(self, distance, longitude, latitude, vlos):
        self.distance = distance * self.KPC_TO_KM
        self.longitude = longitude * m.pi / 180 # deg -> radian
        self.latitude = latitude * m.pi / 180 # deg -> radian
        self.vlos = vlos

    def galvrad(self):
        vr = self.distance * self.vlos * m.cos(self.latitude)**2
        vr += (self.VCIRC * self.sinphi + self.U) * (self.galr * self.cosphi + self.RSUN)
        vr -= (self.VCIRC * self.cosphi - self.V + self.VCIRC) * self.galr * self.sinphi
        vr /= self.galr + self.RSUN * self.cosphi
        return vr

    @property
    def galx(self):
        r"""Galactocentric x-coordinate."""
        return self.distance * m.cos(self.longitude) * m.cos(self.latitude) + self.RSUN

    @property
    def galy(self):
        r"""Galactocentric y-coordinate."""
        return self.distance * m.sin(self.longitude) * m.cos(self.latitude)

    @property
    def galr(self):
        r"""Galactocentric radius."""
        return m.sqrt(self.galx**2 + self.galy**2)

    @property
    def cosphi(self):
        return self.galx / self.galr

    @property
    def sinphi(self):
        return self.galy / self.galr
```

```python
print(hi_cloud(2.6, 129.1, -2.1, -50).galvrad())
print(hi_cloud(1.7, 145, -1.2, -35).galvrad())
print(hi_cloud(1.4, 148, 0.8, -25).galvrad())
```

```python
raw = np.genfromtxt("../data/frail_weisberg.dat")
sample = {}
sample["l"] = raw[:,0]
sample["b"] = raw[:,1]
sample["v_l"] = raw[:,2]
sample["v_u"] = raw[:,3]
sample["d_l"] = raw[:,4]
sample["d_u"] = raw[:,5]
sample = vice.dataframe(sample)
print(sample)
```

```python
vrad = []
for i in range(len(sample["l"])):
    if not m.isnan(sample["v_l"][i]):
        v = hi_cloud(sample["d_l"][i], sample["l"][i], sample["b"][i], sample["v_l"][i]).galvrad()
        if abs(v) < 25: vrad.append(v)
for v in vrad:
    print("%.2f km/s" % (v))
print("mean velocity: %.2f" % (np.mean(vrad)))
print("standard deviation: %.2f" % (np.std(vrad)))
```

```python
fig = plt.figure(figsize = (3.5, 3.5))
ax = fig.add_subplot(111)
ax.set_xlabel(r"$v_{r,g}$ [km/s]")
ax.set_ylabel(r"Counts")
ax.set_xlim([-18, 18])
ax.set_ylim([0, 8])
ax.set_title(r"HI Pulsar Kinematics")

n = 0.3
values = []
for i in range(len(sample["l"])):
    if not m.isnan(sample["v_l"][i]):
        lower = hi_cloud(sample["d_l"][i], sample["l"][i], sample["b"][i], sample["v_l"][i]).galvrad()
        if abs(lower) > 25: continue
    else:
        lower = None
    if not m.isnan(sample["v_u"][i]):
        upper = hi_cloud(sample["d_u"][i], sample["l"][i], sample["b"][i], sample["v_u"][i]).galvrad()
        if abs(upper) > 25: continue
    else:
        upper = None
    kwargs = {"c": named_colors()["black"]}
    if lower is not None and upper is not None:
        # measurement
        mean = (lower + upper) / 2
        values.append(mean)
        ax.errorbar(mean, n, xerr = abs(upper - lower) / 2,
            marker = markers()["point"], capsize = 0, markersize = 10,
            **kwargs)
    elif lower is None or upper is None:
        if lower is None:
            value = upper
        else:
            value = lower
        values.append(value)
        ax.scatter(value, n, marker = markers()["x"], s = 30, **kwargs)
    n += 0.3

counts, bins = np.histogram(values, bins = 20, range = [-6, 6])
ax.plot(bins[:-1], counts, c = named_colors()["crimson"], drawstyle = "steps-mid", zorder = -100)
kwargs = {
    "c": named_colors()["crimson"],
    "fontsize": 14
}
# ax.text(1, 7, r"$\langle v_R \rangle = %.2f$ km/s" % (np.mean(values)), **kwargs)
ax.text(1, 6.8, r"$\sigma = %.2f$ km/s" % (np.std(values)), **kwargs)
print(np.mean(values))
print(np.std(values) / np.sqrt(len(sample["l"])))

plt.tight_layout()
for ext in ["pdf", "jpeg"]:
    kwargs = {}
    if ext == "jpeg": kwargs["dpi"] = 200
    plt.savefig("./empirical/hi-pulsar-measurements.%s" % (ext), **kwargs)
plt.show()
```
